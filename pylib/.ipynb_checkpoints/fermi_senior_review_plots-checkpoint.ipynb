```python
import numpy as np
import matplotlib.pylab as plot
from astropy.io import ascii
from astropy.io import fits
from astropy.table import Table
import datetime
```

```python
def read_bib(dir='/Users/jracusin/Fermi/Senior_Review/SR2019/'):
    file='fermi_bibcodes2.csv'
    bib0=ascii.read(dir+file)
    print bib0.keys()
    date=np.array(bib0['Date Published'])
    d=np.array(np.zeros(len(date)))
#    c=np.array(bib0['No. of Citations'])
    year=np.zeros(len(date))
    month=np.zeros(len(date))
    
    for i in range(len(date)):
        if len(date[i])==8: 
            year[i]=float(date[i][4:8])
            month[i]=float(datetime.datetime.strptime(date[i][0:3], '%b').month)
            d[i]=year[i]+month[i]/12.

    w=np.where(d>0)[0]
    d=d[w]
    year=year[w]
    month=month[w]
#    citations=c[w]
    return d,year,month#,citations
```

```python
date,year,month=read_bib()
```

```python
corr=1./(max(date)-max(year))
```

```python
plot.figure()
h=np.histogram(year,bins=np.linspace(2005.5,2018.5,14))
#h[0][-1]=h[0][-1]*corr
plot.bar(h[1][:-1]+0.5,h[0])
plot.xlabel('Year of Publication')
plot.ylabel('Number of Fermi Papers')
plot.savefig('Fermi_publications.pdf')
plot.show()

plot.figure()
s=np.argsort(date)
plot.plot(date[s],np.cumsum(np.ones(len(year[s]))))
plot.xlabel('Year of Publication')
plot.ylabel('Cumulative Number of Fermi Papers')
plot.show()
```

```python
### MW/MM plot - ugh

t=Table(rows=[('Fermi',2008.5,2020,2025,8e3,300e9,'red','salmon')],\
        names=('Mission','StartYear','StopYear','ExtendedYear','EnergyLow','EnergyHigh','color','ecolor'))
## create object or table with mission name, year, extended years, color, energy range, 
```

```python
plot.figure()
for i in range(len(t)):
    width=(np.log10(t[i]['EnergyHigh'])-np.log10(t[i]['EnergyLow']))
    m=np.log10(t[i]['EnergyHigh'])-width/2.
    minyear=2016
    plot.bar(m,t[i]['StopYear']-minyear,color=t[i]['color'],width=width,bottom=minyear)
    plot.annotate(t[i]['Mission'],(m-1.5,t[i]['StopYear']-1))
    if t[i]['ExtendedYear']>t[i]['StopYear']:
        plot.bar(m,t[i]['ExtendedYear']-minyear,color=t[i]['ecolor'],width=width,bottom=t[i]['StopYear'])
        

plot.xlim([-8,25])
plot.ylim([2016,2025])
#plot.xscale('log')
plot.show()
```
