```python
detalt=np.radians(0.)
detaz=np.radians(0.)
testalt=np.radians(45.)
testaz=np.radians(60.)

def polar2cart(theta,phi):
    return np.array([np.sin(theta) * np.cos(phi),np.sin(theta) * np.sin(phi),np.cos(theta)])

def cart2polar(x,y,z):
#    return np.array([np.arccos(z/np.sqrt(x**2+y**2+z**2)),np.arcsin(y/(np.sqrt(x**2+y**2)))])
#    return np.array([np.arctan(np.sqrt(x**2+y**2)/z),np.arctan(y/x)])
    theta,phi=np.array([np.arccos(z/np.sqrt(x**2+y**2+z**2)),np.arctan(y/x)])
    if np.isnan(theta): theta=0.
    if np.isnan(phi): phi=0.
    return theta,phi
    
x,y,z=polar2cart(testalt,testaz)

px=np.matrix([[1,0,0],[0,np.cos(detalt),np.sin(detalt)],[0,-np.sin(detalt),np.cos(detalt)]])
py=np.matrix([[np.cos(detaz),0,np.sin(detaz)],[0,1,0],[np.sin(detaz),0,np.cos(detaz)]])
pz=np.matrix([[np.cos(detaz),np.sin(detaz),0],[-np.sin(detaz),np.cos(detaz),0],[0,0,1]])

tm=np.matrix([[x,y,z]])
tx,ty,tz=np.matmul(tm,px),np.matmul(tm,py),np.matmul(tm,pz)
print tx,ty,tz
#t,p=cart2polar(tx[0],ty[0],tz[0])
#print np.degrees(t),np.degrees(p)
```

```python
def rotation_separation(detaz,detalt,testaz,testalt):
    ### test point unit vector
    a=polar2cart(testalt,testaz)
    ### detector unit vector
    k=polar2cart(detalt,detaz)
    ## rotate by -az & -alt
    phi=-detaz
    theta=-detalt
#     print a
#     print k
    if phi != 0:
        ap=[a[0]*np.cos(phi)-a[1]*np.sin(phi),a[0]*np.sin(phi)+a[1]*np.cos(phi),a[2]]
        kp=[k[0]*np.cos(phi)-k[1]*np.sin(phi),k[0]*np.sin(phi)+k[1]*np.cos(phi),k[2]]
    else: 
        ap=a
        kp=k
    if theta != 0:
        ap2=[ap[2]*np.sin(theta)+ap[0]*np.cos(theta),ap[1],ap[2]*np.cos(theta)-ap[0]*np.sin(theta)]
        kp2=[kp[2]*np.sin(theta)+kp[0]*np.cos(theta),kp[1],kp[2]*np.cos(theta)-kp[0]*np.sin(theta)]
    else:
        ap2=ap
        kp2=kp

    ap3=cart2polar(ap2[0],ap2[1],ap2[2])
    kp3=cart2polar(kp2[0],kp2[1],kp2[2])
    sep=ap3[0]#eph.separation((0,ap3[0]),(kp3[1],kp3[0]))
#    print np.degrees(kp3)
#    print np.degrees(ap3)
    return sep
```
