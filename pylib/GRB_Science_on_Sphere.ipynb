```python
from astropy.io import fits,ascii
import numpy as np
from astropy import units as u
from astropy.coordinates import SkyCoord
from astropy.table import Table
```

```python
gbm=fits.open('/Users/jracusin/GBM/gbmgrbcat.fits')[1].data
```

```python
n=len(gbm['NAME'])
s=np.argsort(gbm['NAME'])
c=SkyCoord(ra=gbm['RA']*u.deg,dec=gbm['DEC']*u.deg)
```
